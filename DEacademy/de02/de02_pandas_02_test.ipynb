## 數據結構解析

這個JSON數據包括了一個level#1的鍵`shows`映射到一個物件的列表:

![](images/tv_json_01.png)

每一個在`shows`的物件也是一個dict物件, 包括了以下的鍵值對(key-value pairs):
* **show** - 影集名稱
* **runtime** - 單集播放時間長度（分鐘)
* **network** - 頻道名稱
* **episodes** - 影集集數列表

![](images/tv_json_02.png)

由於同一個影集大多數會由多個集數來組成, 因此`episodes`的值也是一個物件的列表:
* **season** - 季
* **episode** - 單集編號
* **name** - 單集的標題
* **air_date** - 播放日期

讓我們預看一下第一個Column的Cell裡頭的資料:

```python
# 從 dataframe 中取出第一筆
first_obj = df_tv_shows.loc[0, "shows"]

# 檢查物件的類型
print(type(first_obj))

# 打印 dict 物件的 keys
print(first_obj.keys())

# 引入pprint的套件來幫助打印 human readable的格式
import pprint

print("")
pprint.pprint(first_obj)
```
