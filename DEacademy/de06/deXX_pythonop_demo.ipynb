```python
from airflow import DAG
from airflow.operators.python import PythonOperator
from airflow.utils.dates import days_ago

default_args = {
    'owner':'EMPLOYEE_ID', # owner是DAG的開發者, 例如: 員工8703147
}


def print_message(task_name):
    print(f'Task name: {task_name}')


dag = DAG(
    dag_id="deXX_pythonop_demo", # prefix必需是tenant id, 例如: de00
    description="dag to demo python operator",
    start_date=days_ago(2),
    schedule_interval=None,
    catchup=False,
    default_args=default_args,
    access_control={
        'TENANT_ID': {'can_read', 'can_edit'} # 設定DAG歸屬那個團隊[tenant id]與權限
    }
)

task01 = PythonOperator(
    task_id='task01',
    python_callable=print_message,
    op_kwargs={'task_name': 'tak01'},
    dag=dag,
)

task02 = PythonOperator(
    task_id='task02',
    python_callable=print_message,
    op_kwargs={'task_name': 'tak02'},
    dag=dag,
)

task03 = PythonOperator(
    task_id='task03',
    python_callable=print_message,
    op_kwargs={'task_name': 'tak03'},
    dag=dag,
)

task04 = PythonOperator(
    task_id='task04',
    python_callable=print_message,
    op_kwargs={'task_name': 'tak04'},
    dag=dag,
)

# Set dependencies between all tasks
task01 >> [task02, task03] >> task04
```
